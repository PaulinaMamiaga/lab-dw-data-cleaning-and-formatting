## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
# Your code here

# Challenge 1: Data Cleaning and Formatting
# Exercise 2: Cleaning invalid  values

# Objective:
# - Identify and handle invalid values in the dataset
# clean these columns with inconsistent and incorrect values to ensure that they only contain valid data
# focus only on cleaning inconsistent values and not involve handling null values (NaN or None).
# standardized Gender column to "M" and "F"
# replaced State abbreviations with its full name or example "AZ": "Arizona", "Cali": "California", "WA": "Washington"
# replace bachelors by Bachelor in education column
# In Customer Lifetime Value, delete the `%` character
# In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"

import pandas as pd

# Load the dataset
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
data = pd.read_csv(url)

# Display original data
print("Original data (first rows):")
print(data.head())

# Standardize column names:
# - convert to lowercase
# - replace spaces with underscores
data.columns = (
    data.columns
        .str.lower()
        .str.replace(" ", "_")
)

# - replace 'st' ONLY if the column name is exactly 'st'
data.columns = [
    "state" if col == "st" else col
    for col in data.columns
]

# Clean inconsistent values (NO null handling)

# - Standardize gender values
# Note: dataset includes values like "F", "M", "Femal", "Male", "female"
data["gender"] = data["gender"].replace({
    "Male": "M",
    "female": "F",
    "Femal": "F"
})

# - Standardize state abbreviations / variants
data["state"] = data["state"].replace({
    "AZ": "Arizona",
    "Cali": "California",
    "WA": "Washington"
})

# - Standardize education values
data["education"] = data["education"].replace({
    "Bachelors": "Bachelor"
})

# - Remove % from Customer Lifetime Value (keep as string for now if needed)
# Hint says: delete the % character (it does NOT require type conversion here)
data["customer_lifetime_value"] = data["customer_lifetime_value"].str.replace("%", "", regex=False)

# - Standardize vehicle class: group into "Luxury"
data["vehicle_class"] = data["vehicle_class"].replace({
    "Sports Car": "Luxury",
    "Luxury SUV": "Luxury",
    "Luxury Car": "Luxury"
})

# Display cleaned data
print("\nCleaned data (first rows):")
print(data.head())
```

## Exercise 3: Formatting data types

The data types of many columns in the dataset appear to be incorrect. This could impact the accuracy of our analysis. To ensure accurate analysis, we need to correct the data types of these columns. Please update the data types of the columns as appropriate.

It is important to note that this exercise does not involve handling null values (NaN or None).

*Hint*:
- *Customer lifetime value should be numeric*
- *Number of open complaints has an incorrect format. Look at the different values it takes with `unique()` and take the middle value. As an example, 1/5/00 should be 5. Number of open complaints is a string - remember you can use `split()` to deal with it and take the number you need. Finally, since it should be numeric, cast the column to be in its proper type.*

```python
# Your code here

import pandas as pd

# Load the dataset
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
data = pd.read_csv(url)

# Display original data
print("Original data (first rows):")
print(data.head())

# Standardize column names:
# - convert to lowercase
# - replace spaces with underscores
data.columns = (
    data.columns
        .str.lower()
        .str.replace(" ", "_")
)

# - replace 'st' ONLY if the column name is exactly 'st'
data.columns = [
    "state" if col == "st" else col
    for col in data.columns
]

# Clean inconsistent values (NO null handling)

# - Standardize gender values
# Note: dataset includes values like "F", "M", "Femal", "Male", "female"
data["gender"] = data["gender"].replace({
    "Male": "M",
    "female": "F",
    "Femal": "F"
})

# - Standardize state abbreviations / variants
data["state"] = data["state"].replace({
    "AZ": "Arizona",
    "Cali": "California",
    "WA": "Washington"
})

# - Standardize education values
data["education"] = data["education"].replace({
    "Bachelors": "Bachelor"
})

# - Remove % from Customer Lifetime Value (keep as string for now if needed)
# Hint says: delete the % character (it does NOT require type conversion here)
data["customer_lifetime_value"] = data["customer_lifetime_value"].str.replace("%", "", regex=False)

# - Standardize vehicle class: group into "Luxury"
data["vehicle_class"] = data["vehicle_class"].replace({
    "Sports Car": "Luxury",
    "Luxury SUV": "Luxury",
    "Luxury Car": "Luxury"
})

# Format customer_lifetime_value as float
# Remove '%' character and convert to float
data["customer_lifetime_value"] = (
    data["customer_lifetime_value"]
        .str.replace("%", "", regex=False)
        .astype(float)
)

# Format number_of_open_complaints as integer
# Extract the middle value from the format '1/x/00'
data["number_of_open_complaints"] = (
    data["number_of_open_complaints"]
        .str.split("/")
        .str[1]
        .astype(int)
)

# Display data types to verify formatting
print("\nData types after formatting:")
print(data.dtypes)

# Display cleaned data
print("\nCleaned data (first rows):")
print(data.head())
```

## Exercise 4: Dealing with Null values

Identify any columns with null or missing values. Identify how many null values each column has. You can use the `isnull()` function in pandas to find columns with null values.

Decide on a strategy for handling the null values. There are several options, including:

- Fill the null values with a specific value (such as the column mean or median for numerical variables, and mode for categorical variables)
- Fill the null values with the previous or next value in the column

Implement your chosen strategy to handle the null values. You can use the `fillna()` function in pandas to fill null values or `dropna()` function to drop null values.

After formatting data types, as a last step, convert all the numeric variables to integers.

```python
# Your code here

# Identify null values per column
print("Number of null values per column:")
print(data.isnull().sum())

# Handle null values
# Identify numerical columns
numerical_cols = data.select_dtypes(include=["int64", "float64"]).columns

# Fill null values with the median
for col in numerical_cols:
    median_value = data[col].median()
    data[col] = data[col].fillna(median_value)

# Identify categorical columns
categorical_cols = data.select_dtypes(include=["object"]).columns

# Fill null values with the mode
for col in categorical_cols:
    mode_value = data[col].mode()[0]
    data[col] = data[col].fillna(mode_value)

# Display number of null values after handling
print("\nNumber of null values per column after handling:")
print(data.isnull().sum())
```
